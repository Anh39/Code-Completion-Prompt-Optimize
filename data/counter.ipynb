```python
import os
import json
from transformers import AutoModelForCausalLM, AutoTokenizer, Qwen2TokenizerFast
MODEL_ID = "Qwen/Qwen2.5-Coder-0.5B"

REPO = "<|repo_name|>"
FILE = "<|file_sep|>"
EOS = "<|endoftext|>"
tokenizer: Qwen2TokenizerFast = AutoTokenizer.from_pretrained(MODEL_ID, trust_remote_code=True, use_fast=True)
total_intra_file_length = 0
total_cross_file_length = 0
with open("train_data_filtered.jsonl", 'r', encoding='utf-8') as file:
    lines = file.read().splitlines()
    for line in lines:
        if line.strip():
            sample = json.loads(line)
            repo_header = REPO + sample["repo_name"] + "\n"
            cfc = "".join(FILE + item["path"] + "\n" + item["text"] for item in sample["cfc_contexts"])
            header = FILE + sample["path"] + "\n"
            content = sample["content"]

            content_with_eot = content + EOS
            cfc_header = repo_header + cfc + header
            total = cfc_header + content_with_eot
            total_intra_file_length += len(tokenizer.encode(content))
            total_cross_file_length += len(tokenizer.encode(cfc_header))
print(total_intra_file_length)
print(total_cross_file_length)
print(total_intra_file_length+total_cross_file_length)
print(f"{total_intra_file_length/(total_intra_file_length+total_cross_file_length) * 100:.2f}")
```
